```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

1. Top Scorer Data

```python
url = "https://www.bbc.com/sport/football/premier-league/top-scorers"
```

```python
try:
    response = requests.get(url)
    response.raise_for_status()
except Exception as e:
    print(f"Excception Error : {e}")
else:
    soup = BeautifulSoup(response.content, "html.parser")
    print("Successfully GET!")
```

```python
print("\n", soup.find("tbody"))
print("\n", soup.find_all("tbody"))
print("\n Total tbody are : ", len(soup.find_all("tbody")))
print("\n One tbody length are : ", len(soup.find("tbody")))
```

```python
print(soup.find("tbody").find_all("tr", class_ = "ssrcss-1ffchex-TableRowBody e1icz100"))
players = soup.find("tbody").find_all("tr", class_ = "ssrcss-1ffchex-TableRowBody e1icz100")
```

```python
#player_name = []
```

```python
player_name = []
city_name = []
player_goal = []
assists = []
playeded = []
goals_per_90 = []
mpgg = []
goals_conversion = []

for player in players:
    # # Extract players names
    player_na = player.find("div", class_ = "ssrcss-1rqvnx-PlayerName e1n8xy5b1").get_text(strip = True)

    ## Extract city/team info
    city = player.find("div", class_ = "ssrcss-cfavl3-TeamsSummary e1n8xy5b0").get_text(strip = True)

    # Extract players goals
    goal = int(player.find("div", class_ = "ssrcss-1fyor78-CellWrapper ef9ipf0").get_text(strip = True))

    # Extract Assists
    assist = int(player.find("div", class_ = "ssrcss-ero735-CellWrapper ef9ipf0").get_text(strip = True))

    # player data within same class
    multiple_data_stats = player.find_all("div", class_ = "ssrcss-ero735-CellWrapper ef9ipf0")
    index = multiple_data_stats[:10]
    print(index)
    played = int(index[2].get_text(strip = True))
    goal_per_90 = index[4].get_text(strip = True)
    mpg = int(index[5].get_text(strip = True))
    goal_conversion = index[8].get_text(strip = True)
     
    # append to lists
    goals_conversion.append(goal_conversion)
    mpgg.append(mpg)
    goals_per_90.append(goal_per_90)
    playeded.append(played)
    assists.append(assist)
    player_goal.append(goal)
    city_name.append(city)
    player_name.append(player_na)
    
    # break
```

```python
print(player_name)
print(city_name)
print(player_goal)
print(assists)
print(playeded)
print(goals_per_90)
print(mpgg)
print(goals_conversion)
```

```python
# convert to dataframe
data = {
    "Player_Name" : player_name,
    "Team_City" : city_name,
    "Goals" : player_goal,
    "Assists" : assists,
    "Played" : playeded,
    "Goals_Per_90" : goals_per_90,
    "MPG" : mpgg,
    "Goal_Coversion" : goals_conversion
}
```

```python
data_frame = pd.DataFrame(data)
```
